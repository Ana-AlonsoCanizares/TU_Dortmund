```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
y = np.array([10,26,14,7,8])
```

c) Using the Quadratic Loss Function

```python
opt_const = np.mean(y)
print('The optimal constant value applying the arithmetic mean of the responses is: ', opt_const)
```

```python
losses_array = (y-opt_const)**2

table = pd.DataFrame({
    'y_i': y,
    'Loss (y_i - theta*)^2': losses_array
})

print('Table of the y_i and each loss value:\n')
print(table)
```

```python
theta_vals = np.linspace(min(y)-5, max(y)+5, 100)
total_loss = [np.mean((y - theta) ** 2) for theta in theta_vals]

plt.figure(figsize=(8, 5))
plt.plot(theta_vals, total_loss, label='Empirical Risk (Loss)')
plt.axvline(opt_const, color='red', linestyle='--', label=f'Optimal θ* = {opt_const:.2f}')
plt.scatter([opt_const], [np.mean(losses_array)], color='red',marker='o', linewidths=1)
plt.title('Empirical Risk vs Constant θ')
plt.xlabel('θ (Constant Prediction)')
plt.ylabel('Average Quadratic Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
opt_const_abs = np.median(y)
print('The optimal constant value applying the median of the responses is: ', opt_const_abs)
```

```python
losses_array_abs = abs(y-opt_const_abs)

table = pd.DataFrame({
    'y_i': y,
    'Loss abs(y_i - theta*)': losses_array_abs
})

print('Table of the y_i and each loss value:\n')
print(table)
```

```python
theta_vals = np.linspace(min(y)-5, max(y)+5, 400)
total_loss = [np.mean(abs(y - theta)) for theta in theta_vals]

plt.figure(figsize=(8, 5))
plt.plot(theta_vals, total_loss, label='Empirical Risk (Absolute Loss)')
plt.axvline(opt_const_abs, color='red', linestyle='--', label=f'Optimal θ* = {opt_const_abs:.2f}')
plt.scatter([opt_const_abs], [np.mean(losses_array_abs)], color='red',marker='o', linewidths=1)
plt.title('Empirical Risk vs Constant θ (Abs Loss)')
plt.xlabel('θ (Constant Prediction)')
plt.ylabel('Average Absolute Loss')
plt.legend()
plt.grid(True)
plt.show()
```
